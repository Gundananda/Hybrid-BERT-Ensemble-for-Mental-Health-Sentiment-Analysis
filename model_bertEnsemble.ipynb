```python
# ------------------- Import Libraries -------------------
import joblib
import tensorflow_hub as hub
import tensorflow as tf
import numpy as np
import tensorflow_text  # Required for BERT preprocessing

import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

import gradio as gr #for ui
```

```python


# ------------------- Text Preprocessing Functions -------------------
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

def clean_text(text):
    text = text.lower()
    text = re.sub(r"http\S+|www\S+|https\S+", '', text, flags=re.MULTILINE)
    text = re.sub(r'\@w+|\#','', text)
    text = re.sub(r'[^A-Za-z\s]', '', text)
    return text

def further_clean_text(text):
    words = text.split()
    words = [w for w in words if w not in stop_words]
    return ' '.join(words)

def lemmatize_text(text):
    words = text.split()
    words = [lemmatizer.lemmatize(w) for w in words]
    return ' '.join(words)

def preprocess(text):
    text = clean_text(text)
    text = further_clean_text(text)
    text = lemmatize_text(text)
    return text

# ------------------- Load TF-IDF Vectorizer and Stacking Model -------------------
# Ensure these files are uploaded in your Colab session before running!
# You can use the file upload button in the left sidebar to upload them

vectorizer = joblib.load("/content/tfidf_vectorizer.pkl")
stacking_model = joblib.load("/content/stacking_model.pkl")

# ------------------- Load BERT Layers -------------------
bert_preprocess = hub.KerasLayer("https://tfhub.dev/tensorflow/bert_en_uncased_preprocess/3")
bert_encoder = hub.KerasLayer("https://tfhub.dev/tensorflow/bert_en_uncased_L-12_H-768_A-12/4")

def get_bert_embedding(texts):
    preprocessed = bert_preprocess(texts)
    return bert_encoder(preprocessed)['pooled_output'].numpy()

# ------------------- Generate Features -------------------
def generate_features(text):
    cleaned_text = preprocess(text)
    bert_emb = get_bert_embedding([cleaned_text])
    tfidf_feat = vectorizer.transform([cleaned_text]).toarray()
    combined = np.hstack((bert_emb, tfidf_feat))
    return combined

# ------------------- Predict Function -------------------
def predict(text):
    features = generate_features(text)
    pred = stacking_model.predict(features)
    return pred[0]

# ------------------- Class Labels -------------------
label_map = {
    0: "Normal",
    1: "Depression",
    2: "Suicidal",
    3: "Anxiety",
    4: "Bipolar",
    5: "Stress",
    6: "Personality Disorder"
}
```

```python

# ------------------- Gradio Interface Function -------------------
def predict_label(text):
    label_id = predict(text)
    return f"Predicted Class: {label_map[label_id]}"

interface = gr.Interface(
    fn=predict_label,
    inputs=gr.Textbox(lines=4, placeholder="Enter text here...", label="Mental Health Text"),
    outputs=gr.Text(label="Prediction"),
    title="Mental Health Text Classifier",
    description="Enter a sentence to classify into a mental health category (e.g., depression, anxiety, stress, etc.)"
)

```

```python
# ------------------- Launch Interface -------------------
interface.launch(share=True)
```
